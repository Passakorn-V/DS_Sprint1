```python
import pandas as pd
import plotly.express as px

# กำหนดช่วงราคา
price_ranges = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-1000', '1001-2000', '2001-3000', '3001-4000', '4001-5000', '5001-10000', '10000+']

# สร้างกราฟสำหรับแต่ละช่วงราคา
for price_range in price_ranges:
  # กรองข้อมูลเฉพาะข้อมูลในช่วงราคา
  if price_range == '10000+':
    df_price_range = df_pay_true[df_pay_true['price'] >= 10000]
  else:
    price_range_min, price_range_max = map(int, price_range.split('-'))
    df_price_range = df_pay_true[(df_pay_true['price'] >= price_range_min) & (df_pay_true['price'] <= price_range_max)]

  # เรียงข้อมูลจากมากไปน้อย
  df_sorted = df_price_range['day_of_week'].value_counts().sort_values(ascending=False)

  # สร้าง DataFrame จาก Series
  df_bar_data = pd.DataFrame({'day_of_week': df_sorted.index, 'count': df_sorted.values})

  # สร้างกราฟแท่งด้วย Plotly Express
  fig = px.bar(df_bar_data,
                x='day_of_week',  # Set day_of_week as x-axis
                y='count',        # Set count as y-axis
                labels={'day_of_week': 'Day of Week', 'count': 'Count'},  # Set labels for axes
                title=f'Count of Entries by day_of_week (Price Range {price_range})',
                color='day_of_week',  # Use day_of_week for color differentiation
                color_discrete_map=color_map  # กำหนดสีตามค่าที่กำหนดไว้ใน color_map
              )

  # เพิ่ม label ให้กับแกน x และ y
  fig.update_xaxes(title_text='Day of Week')
  fig.update_yaxes(title_text='Count')

  # กำหนดขนาดของกราฟ
  fig.update_layout(width=800, height=600)

  # เพิ่มตัวเลขกำกับในกราฟแต่ละแท่ง
  for bar in fig.data:
    for i, j in enumerate(bar.y):
        text = f'{j}'
        fig.add_annotation(
            x=bar.x[i],
            y=j,
            text=text,
            textangle=0,
            showarrow=False,
            font_size=12,
        )

  # แสดงกราฟ
  fig.show()
```

```python
import pandas as pd
import plotly.express as px

# กำหนดช่วงราคา
price_ranges = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-1000', '1001-2000', '2001-3000', '3001-4000', '4001-5000', '5001-10000', '10000+']

# สร้างกราฟสำหรับแต่ละช่วงราคา
for price_range in price_ranges:
    # กรองข้อมูลเฉพาะข้อมูลในช่วงราคา
    if price_range == '10000+':
        df_price_range = df_pay_true[df_pay_true['price'] >= 10000]
    else:
        price_range_min, price_range_max = map(int, price_range.split('-'))
        df_price_range = df_pay_true[(df_pay_true['price'] >= price_range_min) & (df_pay_true['price'] <= price_range_max)]

    # รวมยอดรวมของราคาตามวันของสัปดาห์
    df_grouped = df_price_range.groupby('day_of_week')['price'].sum().reset_index()

    # เรียงลำดับค่าราคารวมจากมากไปน้อย
    df_grouped_sorted = df_grouped.sort_values(by='price', ascending=False)

    # สร้างกราฟแท่งด้วย Plotly Express
    fig = px.bar(df_grouped_sorted,
                 x='day_of_week',  # ให้แกน x เป็นวันของสัปดาห์
                 y='price',  # ให้แกน y เป็นยอดรวมของราคา
                 labels={'day_of_week': 'Day of Week', 'price': 'Total Price'},  # กำหนด label ให้แกน x และ y
                 title=f'Total Price by day_of_week (Price Range {price_range})',
                 color='day_of_week',  # ใช้วันของสัปดาห์เป็นสี
                 color_discrete_map=color_map  # กำหนดสีตามค่าที่กำหนดไว้ใน color_map
    )

    # เพิ่ม label ให้กับแกน x และ y
    fig.update_xaxes(title_text='Day of Week')
    fig.update_yaxes(title_text='Total Price')

    # กำหนดขนาดของกราฟ
    fig.update_layout(width=800, height=600)

    # เพิ่มตัวเลขกำกับในกราฟแต่ละแท่ง
    for bar in fig.data:
        for i, j in enumerate(bar.y):
            text = f'{j:.0f}'  # แสดงทศนิยม 0 หลัก
            fig.add_annotation(
                x=bar.x[i],
                y=j,
                text=text,
                textangle=0,
                showarrow=False,
                font_size=12,
            )

    # แสดงกราฟ
    fig.show()

```

```python
import pandas as pd
import plotly.express as px

# กำหนดช่วงราคา
price_ranges = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-1000', '1001-2000', '2001-3000', '3001-4000', '4001-5000', '5001-10000', '10000+']

# สร้างกราฟสำหรับแต่ละช่วงราคา
for price_range in price_ranges:
    # กรองข้อมูลเฉพาะข้อมูลในช่วงราคา
    if price_range == '10000+':
        df_price_range = df_pay_true[df_pay_true['price'] >= 10000]
    else:
        price_range_min, price_range_max = map(int, price_range.split('-'))
        df_price_range = df_pay_true[(df_pay_true['price'] >= price_range_min) & (df_pay_true['price'] <= price_range_max)]

    # รวมยอดรวมของราคาตามช่วงเวลา
    df_grouped = df_price_range.groupby('time_range')['price'].sum().reset_index()

    # เรียงลำดับค่าราคารวมจากมากไปน้อย
    df_grouped_sorted = df_grouped.sort_values(by='price', ascending=False)

    # สร้างกราฟแท่งด้วย Plotly Express
    fig = px.bar(df_grouped_sorted,
                 x='time_range',  # ให้แกน x เป็นช่วงเวลา
                 y='price',  # ให้แกน y เป็นยอดรวมของราคา
                 labels={'time_range': 'Time Range', 'price': 'Total Price'},  # กำหนด label ให้แกน x และ y
                 title=f'Total Price by time_range (Price Range {price_range})',
                 color='time_range',  # ใช้ช่วงเวลาเป็นสี
                 color_discrete_map=color_time  # กำหนดสีตามค่าที่กำหนดไว้ใน color_map
    )

    # เพิ่ม label ให้กับแกน x และ y
    fig.update_xaxes(title_text='Time Range')
    fig.update_yaxes(title_text='Total Price')

    # กำหนดขนาดของกราฟ
    fig.update_layout(width=800, height=600)

    # เพิ่มตัวเลขกำกับในกราฟแต่ละแท่ง
    for bar in fig.data:
        for i, j in enumerate(bar.y):
            text = f'{j:.0f}'  # แสดงทศนิยม 0 หลัก
            fig.add_annotation(
                x=bar.x[i],
                y=j,
                text=text,
                textangle=0,
                showarrow=False,
                font_size=12,
            )

    # แสดงกราฟ
    fig.show()
```
